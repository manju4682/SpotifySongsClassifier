```python
final_df = pd.concat([Pop,Rock,Hop,Folk,Metal,Blues,Jazz],axis=0)
```

```python
final_df.info
```

```python
final_df['genre'] = final_df['genre'].replace('Pop',1)
```

```python
final_df['genre'] = final_df['genre'].replace('Rock',2)
final_df['genre'] = final_df['genre'].replace('Hip Hop',3)
final_df['genre'] = final_df['genre'].replace('Folk/Acoustic',4)
final_df['genre'] = final_df['genre'].replace('Metal',6)
final_df['genre'] = final_df['genre'].replace('Blues',8)
final_df['genre'] = final_df['genre'].replace('Jazz',9)
```

```python
final_df['genre'].value_counts()
```

```python
y = final_df.genre
```

```python
final_df.columns
```

```python
features = ['danceability','is_explicit', 'energy', 'key', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness','valence', 'tempo']
```

```python
X = final_df[features]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
